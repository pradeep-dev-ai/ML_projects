```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

#  Scale the data
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

#  Define a base estimator with class weights
base_estimator = DecisionTreeClassifier(
    max_depth=3,
    class_weight='balanced',   # handles imbalance automatically
    random_state=42
)

#  Create AdaBoost model using that base estimator
adaboost = AdaBoostClassifier(
    estimator=base_estimator,   # (use 'base_estimator' if sklearn <1.2)
    n_estimators=150,
    learning_rate=0.8,
    random_state=42
)

# Train the model
adaboost.fit(X_train_scaled, y_train)

#  Predict
y_pred = adaboost.predict(X_test_scaled)

#  Evaluate
print(" Accuracy:", accuracy_score(y_test, y_pred))
print("\n Classification Report:\n", classification_report(y_test, y_pred))

#  Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("AdaBoost Confusion Matrix (Without SMOTE)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

# 1ï¸âƒ£ Scale your data
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train Gradient Boosting model
gb = GradientBoostingClassifier(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=3,
    random_state=42
)
```

```python
#  Scale (optional, but helps for consistency)
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

#  Compute class weight for imbalance
# scale_pos_weight = (Number of negative samples / Number of positive samples)
neg, pos = np.bincount(y_train)
scale_pos_weight = neg / pos

#  Define XGBoost model
xgb_model = XGBClassifier(
    n_estimators=300,          # number of trees
    learning_rate=0.1,         # step size
    max_depth=5,               # depth of each tree
    subsample=0.8,             # % of data for each tree
    colsample_bytree=0.8,      # % of features for each tree
    scale_pos_weight=scale_pos_weight,  # handles imbalance
    eval_metric='logloss',     # evaluation metric
    random_state=42,
    use_label_encoder=False
)

xgb_model.fit(
    X_train_scaled, y_train
)


#  Predict on test set
y_pred = xgb_model.predict(X_test_scaled)

#  Evaluate performance
print("âœ… Accuracy:", accuracy_score(y_test, y_pred))
print("\nðŸ“Š Classification Report:\n", classification_report(y_test, y_pred))

#  Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("XGBoost Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
#  Imports
from lightgbm import LGBMClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#  Scale your data (optional but consistent)
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

#  Compute class imbalance ratio
neg, pos = np.bincount(y_train)
scale_pos_weight = neg / pos
print(f"Class Imbalance Ratio (scale_pos_weight): {scale_pos_weight:.2f}")

#  Initialize LightGBM classifier
lgb_model = LGBMClassifier(
    n_estimators=300,          # Number of boosting rounds
    learning_rate=0.1,         # Step size shrinkage
    max_depth=-1,              # No limit by default
    num_leaves=31,             # Controls complexity (like max_depth)
    subsample=0.8,             # Fraction of data for each tree
    colsample_bytree=0.8,      # Fraction of features for each tree
    scale_pos_weight=scale_pos_weight,  # Handles imbalance
    random_state=42
)

#  Train model
lgb_model.fit(X_train_scaled, y_train)

#  Predictions
y_pred = lgb_model.predict(X_test_scaled)

#  Evaluation
print(" Accuracy:", accuracy_score(y_test, y_pred))
print(" Classification Report:\n", classification_report(y_test, y_pred))

#  Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("LightGBM Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
